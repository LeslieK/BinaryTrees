### Binary Search Tree: keys = range(10000) in random order

```python
from bst import BST
trials = 10000
maxval = 20
b = BST(trials, maxval)

keys = range(trials)
import random
random.shuffle(keys)
for k in keys:
    b.put(k, k)
b.a.plotData()
```

```python
m = b.a.means[10000 - 1]
print m, 1.39 * math.log(10000, 2)
```

### Red-Black Search Tree: keys = range(10000) in random order

```python
from redblacktrees2 import RedBlackBST
trials = 10000
maxval = 20
r = RedBlackBST(trials, maxval)

keys = range(trials)
random.shuffle(keys)

for k in keys:
    r.put(k, k)
    
print r.size()
r.a.plotData()
```

```python
m = r.a.means[trials - 1]
print m, math.log(trials - 1, 2) - .5
r.a.vals[trials - 1]
```

```python
from bst import BST
trials = 13525
maxval = 20
b = BST(trials, maxval)
with open("TaleOfTwoCities.txt") as f:
        words = (word for line in f for word in line.split() if len(word) > 7)
        
        for w in words:
            if w in b:
                b.put(w, b.get(w) + 1)
            else:
                b.put(w, 1)
print b.size()
b.a.plotData()
```

```python
m = b.a.means[13524]
print m, 1.39 * math.log(13524, 2)
```
