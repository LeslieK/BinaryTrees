### Red-Black Trees: plot of height as keys are added in ascending order

```python
from redblacktrees2 import RedBlackBST
trials = 10000
maxval = 20
title = "red black trees (height)"
r = RedBlackBST(trials, maxval, title)

keys = range(trials)

for k in keys:
    r.put(k, k)
    
print r.size(), "max value = ", r.a.vals[trials - 1], "max mean = ", r.a.means[trials - 1]
r.a.plotData()
```
